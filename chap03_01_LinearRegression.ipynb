```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
torch.manual_seed(1)
```

```python
x_train = torch.FloatTensor([[1],[2],[3],[4],[5],[6],[7],[8],[9]])
y_train = torch.FloatTensor([[6],[9],[11],[13],[15],[17],[19],[21],[23]])
print(x_train)
print(x_train.shape)
print(y_train)
print(y_train.shape)
```

```python
#기중치 초기화
W = torch.zeros(1, requires_grad=True)
print(W)
b = torch.zeros(1, requires_grad=True)
print(b)
```

```python
hypothesis = x_train*W + b
print(hypothesis)
```

```python
cost = torch.mean((hypothesis - y_train)**2)
print(cost)
```

##경사 하강법 구현하기

```python
optimizer = optim.SGD([W,b], lr=0.01)
```

```python
nb_epochs = 100

for epoch in range(nb_epochs + 1):

  hypothesis = x_train*W + b

  cost = torch.mean((hypothesis -y_train)**2)

  optimizer.zero_grad()
  print("W,b 1: ",W,b,epoch)  
  cost.backward()
  print("W,b 2: ",W,b,epoch)  
  optimizer.step()
  print("W,b 3: ",W,b,epoch)

  if epoch % 100 == 0:
    print('Epoch {:4d}/{} W: {:.3f}, b: {:.3f} Cost: {:.6f}'.format(
        epoch, nb_epochs, W.item(), b.item(), cost.item()
    ))
  print("------------------")
```
